## Exercise 4 — Counting Errors with a Loop

**Goal:** Use a for loop to count errors across multiple predictions.

**Instructions:**
1. The two lists below represent 10 model predictions and their ground truths.
2. Write a for loop that compares each pair and counts:
   - `correct` — how many match
   - `false_positives` — how many times model said PERMITTED but truth is DENIED
   - `false_negatives` — how many times model said DENIED but truth is PERMITTED
3. Print all three counts and the accuracy as a percentage

````python
# Exercise 4
predictions = [
    "PERMITTED", "DENIED", "PERMITTED", "DENIED", "PERMITTED",
    "PERMITTED", "DENIED",  "DENIED",   "PERMITTED", "DENIED"
]

ground_truths = [
    "PERMITTED", "PERMITTED", "PERMITTED", "DENIED", "DENIED",
    "PERMITTED", "DENIED",    "PERMITTED",  "PERMITTED", "DENIED"
]

correct = 0
false_positives = 0
false_negatives = 0

# Your for loop here
for pred, gt in zip(predictions, ground_truths):
    if pred=="PERMITTED" and gt=="DENIED":   #false positive
        false_positives += 1
    elif pred=='DENIED' and gt=='PERMITTED': #false negative
        false_negatives += 1
    else:                                    #all else (correct)
        correct += 1

# Print results
print(f"Correct:         {correct}")
print(f"False positives: {false_positives}  (PHI leak risk)")
print(f"False negatives: {false_negatives}")
print(f"Accuracy:        {correct / len(predictions):.1%}")
````

## Exercise 5 — Write an accuracy() Function

**Goal:** Package your loop logic into a reusable function.

**Instructions:**
1. Write a function called `accuracy` that:
   - Takes two arguments: `predicted` (a list) and `ground_truth` (a list)
   - Returns the fraction of predictions that are correct (a float between 0.0 and 1.0)
2. Test it on the lists from Exercise 4
3. Also test it on a perfect prediction (all correct) and a zero prediction (all wrong)

**Hint:** You can use `sum()` with a generator expression, or a loop — either is fine.

````python
# Exercise 5
def accuracy(predicted, ground_truth):
    #"""Return the fraction of predictions matching ground_truth."""
    # Your code here
    correct = 0
    false_positives = 0
    false_negatives = 0
    
    for preds, gts in zip(predicted, ground_truth):
        if preds=="PERMITTED" and gts=="DENIED":   #false positive
            false_positives += 1
        elif preds=='DENIED' and gts=='PERMITTED': #false negative
            false_negatives += 1
        else:                                      #all else (correct)
            correct += 1
    fraction = correct / len(predicted)
    return fraction

# Test 1: predictions from Exercise 4
preds = ["PERMITTED","DENIED","PERMITTED","DENIED","PERMITTED",
         "PERMITTED","DENIED","DENIED","PERMITTED","DENIED"]
gts   = ["PERMITTED","PERMITTED","PERMITTED","DENIED","DENIED",
         "PERMITTED","DENIED","PERMITTED","PERMITTED","DENIED"]

print(f"Exercise 4 accuracy: {accuracy(preds, gts):.1%}")

# Test 2: perfect predictions
perfect = ["PERMITTED", "DENIED", "PERMITTED"]
print(f"Perfect accuracy:    {accuracy(perfect, perfect):.1%}")

# Test 3: all wrong
all_wrong_pred = ["DENIED",   "PERMITTED"]
all_wrong_gt   = ["PERMITTED", "DENIED"]
print(f"All-wrong accuracy:  {accuracy(all_wrong_pred, all_wrong_gt):.1%}")
````
