## Exercise 5 — Write an accuracy() Function

**Goal:** Package your loop logic into a reusable function.

**Instructions:**
1. Write a function called `accuracy` that:
   - Takes two arguments: `predicted` (a list) and `ground_truth` (a list)
   - Returns the fraction of predictions that are correct (a float between 0.0 and 1.0)
2. Test it on the lists from Exercise 4
3. Also test it on a perfect prediction (all correct) and a zero prediction (all wrong)

**Hint:** You can use `sum()` with a generator expression, or a loop — either is fine.

````python
# Exercise 5
def compute_accuracy(predicted, ground_truth):
    """Return the fraction of predictions that match ground truth."""
    correct = sum(p == g for p, g in zip(predicted, ground_truth))
    return correct / len(predicted)


# Use it:
preds = ["PERMITTED", "DENIED", "PERMITTED", "DENIED", "PERMITTED"]
gts   = ["PERMITTED", "PERMITTED", "PERMITTED", "DENIED", "PERMITTED"]

acc = compute_accuracy(preds, gts)
print(f"Accuracy: {acc:.1%}")   # 80.0%


# Test 1: predictions from Exercise 4
preds = ["PERMITTED","DENIED","PERMITTED","DENIED","PERMITTED",
         "PERMITTED","DENIED","DENIED","PERMITTED","DENIED"]
gts   = ["PERMITTED","PERMITTED","PERMITTED","DENIED","DENIED",
         "PERMITTED","DENIED","PERMITTED","PERMITTED","DENIED"]
    

print(f"Exercise 4 accuracy: {compute_accuracy(preds, gts):.1%}")

# Test 2: perfect predictions
perfect = ["PERMITTED", "DENIED", "PERMITTED"]
print(f"Perfect accuracy:    {compute_accuracy(perfect, perfect):.1%}")

# Test 3: all wrong
all_wrong_pred = ["DENIED",   "PERMITTED"]
all_wrong_gt   = ["PERMITTED", "DENIED"]
print(f"All-wrong accuracy:  {compute_accuracy(all_wrong_pred, all_wrong_gt):.1%}") 
````
